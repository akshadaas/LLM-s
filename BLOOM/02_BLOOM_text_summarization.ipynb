```python
# from transformers import AutoTokenizer, BloomModel
from transformers import AutoTokenizer, BloomForCausalLM
from accelerate import Accelerator
import torch
```

```python
# Define the BLOOM model name
model_name = "bigscience/bloom-560m"
```

```python
# Initialize the accelerator
# accelerator = Accelerator()
```

```python
# Initialize the tokenizer and model
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = BloomForCausalLM.from_pretrained(model_name)
```

```python
# device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
# model = model.to(device)
```

```python
# Use the accelerator to move the model to the GPU (if available)
# model, tokenizer = accelerator.prepare(model, tokenizer)
```

```python
# Sample text for summarization
text = "The European Union has extended its Covid-19 vaccine export controls to include countries with high vaccination rates that have not reciprocated the measures, as concerns grow about the threat of the Delta variant. The updated regulation came into force on 29 June, and covers vaccine exports from the EU to 17 countries, including Saudi Arabia, Canada and Japan, as well as the UK, which is no longer part of the bloc. Since January, the EU has had the power to block vaccine exports if manufacturers failed to meet contractual obligations to the bloc."
```

```python
# Tokenize the input text
inputs = tokenizer(text, return_tensors="pt")
```

```python
# Move the inputs to the GPU (if available)
# inputs = accelerator.prepare(inputs)
```

```python
# Generate the summary
summary_ids = model.generate(
    inputs["input_ids"],
    attention_mask=inputs["attention_mask"],
    max_length=110,
    min_length=30,
    num_beams=6,
    early_stopping=True
)
```

```python
# Decode the summary
summary = tokenizer.decode(summary_ids[0], skip_special_tokens=True)

# Print the summary
print(summary)
```

```python
# Importing required libraries
import torch
from transformers import T5Tokenizer, T5ForConditionalGeneration
```

```python
model_name="t5-small"
```

```python
tokenizer = T5Tokenizer.from_pretrained(model_name)
model = T5ForConditionalGeneration.from_pretrained(model_name)
```

```python
# Sample text for summarization
input_text = "The European Union has extended its Covid-19 vaccine export controls to include countries with high vaccination rates that have not reciprocated the measures, as concerns grow about the threat of the Delta variant. The updated regulation came into force on 29 June, and covers vaccine exports from the EU to 17 countries, including Saudi Arabia, Canada and Japan, as well as the UK, which is no longer part of the bloc. Since January, the EU has had the power to block vaccine exports if manufacturers failed to meet contractual obligations to the bloc."
```

```python
input_text = "summarize: " + input_text
input_tokens = tokenizer.encode(input_text, return_tensors="pt", max_length=512, truncation=True)
```

```python
summary_ids = model.generate(input_tokens, max_length=100, num_return_sequences=1, early_stopping=True)
summary = tokenizer.decode(summary_ids[0], skip_special_tokens=True)
```

```python
print(summary)
```
